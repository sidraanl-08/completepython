## 1. What is a compiler?

- A compiler is a special program that translates the entire source code of a programming language into machine code (or an intermediate form) before the program is run

- Input: Human-readable source code (like C, C++, or Java).

- Output: Machine code or an executable file that the computer can run directly.

- When errors are found: The compiler shows errors before execution (compile-time errors).

- Why we use it: It makes the program fast because the translation is done once, and the executable can run multiple times without translating again.

### Analogy:

- Think of a compiler as a translator who converts an entire book from English to French before anyone reads it. You don’t need to translate each sentence while reading; it’s all ready to read at once.

## 2. What is a Compiled Language?

- It is a language in which where the program once compiled is expressed in the instructions of the target machine. The machine code is undecipherable by humans. 

### Common Types:

- C/C++/C#
- CLEO
- COBOL

### Analogy:

- C: Like writing the full recipe, translating it into a ready-made meal kit, and then cooking without checking each step while cooking.



## 3. What is an Interpreted Language?

- It is generally interpreted without compiling the program into machine instructions. It is the one where the instructions are not directly executed by target machine. But instead ready and executed by some other program 

### Common Types:

- Python 
- Javascript
- Perl
- BASIC 

### Analogy:

- Python: Like reading a recipe and cooking step by step, fixing mistakes as you go.
