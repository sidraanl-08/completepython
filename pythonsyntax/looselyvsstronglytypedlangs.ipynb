## 1. What is a Loosely Typed Language?

- It is a language where data types are loosely, and casually defined 

- More flexible; can mix types, often auto-converts

- Type errors may go unnoticed, can cause unexpected results


### Examples: 

- JavaScript

- PHP

- Pearl 

### Code Example:

let x = "5";
let y = 2;

// JavaScript automatically converts types
let z = x + y;
console.log(z);  // "52"


## 1. What is a Strongly Typed Language?

- It is a language where data types are properly define 

- Enforces strict rules; cannot mix types without explicit conversion

- Type errors are caught immediately (compile-time or runtime)


### Examples: 

- Python

- Java

- C# 

### Code Example:

x = "5"
y = 2

This will give an error because Python won't automatically convert types
z = x + y  

TypeError: can only concatenate str (not "int") to str

we must do type casting explicitly

z = int(x) + y  # now it works
print(z)  # 7
