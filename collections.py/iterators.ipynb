# What Are Python Iterators?

- Iterators are methods that iterate collections like lists, tuples, etc. Using an iterator method, we can loop through an object and return its elements.



# What Is Iterator Protocol?

- Technically, a Python iterator object must implement two special methods, __iter__() and __next__(), collectively called the iterator protocol.

```python
#iterating through an iterator with iter() and next()

mylist = [4,7,0]

iterator = iter(mylist) #the iter() is the iterator to be used with next()

print(next(iterator)) #next gives the next() element present in the index 
print(next(iterator))
print(next(iterator))




#using for loop with iterator

mylist2 = [13,45,67,23,1,56,34]

iterator2 = iter(mylist2)

for i in iterator2:
    print(i)
```
