## Understanding exception/error handling in python

### What is exception handling in python?

- In simple terms, python has some built-in exceptions that are raised when our code undergoes some errors

- When these exceptions occur, the python interpretor stops the current process and passes it to calling process untill these exceptions are handled 

- If these exceptions are not handled, the program will just crash 



### Common Types Of Exception Handling Are:

# 1. try-except
# 2. try-except-else
# 3. try-except-else-finally


```python
#use case of try-except error handling

try: 
    numerator = 10
    denominator = 0
    result = numerator/denominator 

    print(result) #this will raise an exception coz we cannot divide a number by 0 in python

except: #this except will catch this exception and the code inside it will be run
    print("error! denominator cannot be zero")



#there can be multiple except blocks after try block since they help us manage different errors easily 

try:
    
    even_numbers = [2,4,6,8]
    print(even_numbers[5])

except ZeroDivisionError: #this except block is not run coz it does not match the type of exception we have raised
    print("denominator cannot be zeroo")

except IndexError: #since this except block matches the exception, so its used 
    print("index out of bound/range")
```

```python
#use case of try-except-else error handling 

try: #this is the risky/doubtful 
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError: #this handles 1 type of error
    print("Error: You cannot divide by zero!")
except ValueError: #this handles another type of error
    print("Error: Please enter a valid number.")
else: #this will give the right output when no error occurs 
    print("Success! The result is:", result)
```

```python
#use case of try-except-else-finally
#we can use as many except blocks but there can be only 1 finally block for try 

try:
    numerator = 10
    denominator = 0

    result = numerator/denominator

    print(result)

except: 
    print("Error: Denominator cannot be 0.")

    
finally: #finally will always run no matter there is an exception or not
    print("This is finally block.")
```
