## 1. What are python virtual environments?

- At its core, the main purpose of Python virtual environments is to create an isolated environment 
for Python projects. This means that each project can have its own dependencies, regardless of 
what dependencies every other project has. 

## 2. What are python packages?

- A Python package contains all the files you need for a module.


## 3. What are python modules?

- Modules are Python code libraries you can include in your project.
