2. Dictionary counting / frequency map: Mental shortcut: “If I need counts → use dict with key=element, value=count, increment if exists, else initialize.”

5. Adjacent element checking (index-based loop), Boolean flag + early exit (break) : Mental shortcut: “If checking condition over pairs → loop with range(len(list)-1) and check list[i] + list[i+1]. Stop early if found.”

```python

nums = [3, 7, 2, 9, 4, 8, 1]

newnums = [num for num in nums if num > 5]
print(newnums)


#Q:2 Create a dictionary that stores each character and how many times it occurs

text = "mississippi"

characts = {}

for char in text:
    if char in characts:
        characts[char] +=1
    else:
        characts[char] = 1


print(characts)


#Q:3 Write a program that prints the index and value of each element in a list

letters = ['a', 'b', 'c', 'd']

for index, value in enumerate(letters):
    print(index,value)


#Q:4 Find the highest temperature without using max().

temps = [21, 19, 23, 18, 25, 29, 17]

currentmax = temps[0]

for temp in temps:
    if temp > currentmax:
        currentmax = temp

print(currentmax)


#Q:5 Check if there exists any two adjacent numbers whose sum is greater than 10. Print True or False.

points = [2, 5, 1, 8, 6, 3]

for i in range(len(points)-1):
    if points[i] + points[i+1] > 10:
        found = True
        break

print(found)
```

# 2. Python Practice Session - Day2

## Patterns Practiced:

- Value − Index → subtract the index from the value.

- Even/Odd index operations → multiply or transform based on index parity.

- Cumulative sum / “growing window” → sum elements from start to current index.

- Compare with previous element → track changes across the list.

- Reverse + Index operations → combine list reversal with index-based changes.

```python
#Q:1 Create a new list where each element = value − index

nums = [4, 9, 2, 7, 5]

result = []

for i, val in enumerate(nums):
    result.append(val - i)

print(result)



#Q:2 Create a list where: Even indices → multiply value by 2, Odd indices → multiply value by 3

data = [2, 3, 5, 7, 11, 13]

result2 = []

for i, val in enumerate(data):
    if i % 2 == 0:
        result2.append(val*2)
    else:
        result2.append(val*3)
print(result2)



#Q:3 "Growing Window" Sum Pattern

arr = [1, 2, 3, 4, 5]

result3 = []
running = 0

for val in arr:
    running += val
    result3.append(running)
    
print(result3)


#Q:4 Count Values Following Condition Pattern: value > previous value (except first index)

temps = [11, 14, 17, 21, 25, 19, 15]

count = 0 

for i, val in enumerate(temps):
    if i == 0:
        continue
    if val > temps[i-1]:
        count +=1
print(count)



#Q:5 Create a new list where the list is reversed and each value is increased by its new index.

nums2 = [10, 20, 30, 40]

newnums = []

revlist = nums2[::-1]

for i, val in enumerate(revlist):
    newnums.append(val + i)
print(newnums)
    
```
