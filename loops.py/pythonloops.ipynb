## 1. While Loop

```python
#first the value of i was 1
#loop will only work if the condition is True
#it checks if 1 is less than 6 so it gives 1
#but 1 has +1 so it becomes 2
#now it goes back to the condition given and checks if 2 is less than 6
#the condition is True, so it will print 2 and so on 
#in the end, 6 is not less than 6 means condition is not satisfied
#as soon as the condition is not satisfied, the loop will stop 
#we must make the condition in while loop False at some point, otherwise it will crete infinite loop 

i = 1

while(i<6):
    print(i)
    i+=1

#creating a while loop to print numbers 1-10
#here we used 11 coz 11<11 is False, and the loop stops 

i = 1

while(i<11):
    print(i)
    i+=1

#printing contents of a list using while loop

l = [100, "sid", False, "this", "amsterdam", "city", "nl"]

i = 0 

#here it means that keep printing untill i is less than length of list which is 7

while(i<len(l)):
   print(l[i])
   i+=1

```

## 2. For Loop

- For loops are great for sequential traversing like lists, tuples, and strings

```python
#even if we dont add 0, python will start looping from 0 and skip 6
#so it prints only till 5

for i in range(6):
    print(i)

#for loop works in lists 
#listA

list = [1,4,6,234,6,764]

for i in list:
    print(i)

#listB

list1 = [1,2,3,4,5]

for num in list1:
    print(num)

#for loop works in tuples

t = (96,231,54,89,100)

for i in t: #this is called linear search 
    print(i)

#for loop works in strings also

string = "sidra"

for i in string:
    print(i)
```

### 1. Using For Loop With Else


```python
#for loop will keep looping untill the end
#but once it reaches the end, it will print else statement

list2 = [1,4,7]

for i in list2:
    print(i)
else:
    print("i dont have any values left, so i cannot loop anymore")


str = "amsterdam"

for char in str:
    print(char)
else:
    print("loop ended")
```

### 2. Break, Continue and Pass in For loops

- break will stop the for loop immediately

- continue will skip the iteration given and jump onto the next one

- pass is a null statement in python loops, it tells them to do nothing

```python
#as soon as it checks i == 3 it will come out of the loop, it stops immediately no matter what

for i in range(5):
    if (i == 3):
        break
    print(i)

#as soon as it checks i == 4 it will skip 4 and jump onto 5
#it prints only till 6 coz range stops at 7, it wont print 7

for i in range(7):
    if (i == 4):
        continue
    print(i)


#pass tells python that dont throw an error, we will work on this loop later
#without pass, it will throw an error

group = [1,3,5]

for i in group:
    pass
```

## 3. range() In Python Loops

- we can use range function to get a sequence of numbers
- it starts from 0 by default 
- it increases from 1 by default
- it stops before a specified number
- start, step size, end 

```python
seq = range(5)

for i in seq:
    print(i)

#works same with range() also
seq2 = range(5)
print(i)

for i in seq2:
    print(i)
    
#giving 2 arguments in range() means giving a start and stop
seq2 = range(2, 10)

for i in seq2:
    print(i)
    

#step size works as increment in range()
seq2 = range(2,10,2) #2 means create step size of 2 means create gap of 2

for i in seq2:
    print(i)
    
```
